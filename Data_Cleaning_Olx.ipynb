```python
for i in range(df.shape[0]):
    df['Car Price'].loc[i] = (df['Car Price'].loc[i]).replace('₹','')
    df['Car Price'].loc[i] = (df['Car Price'].loc[i]).replace(',','')
    df['Car Price'].loc[i] = int(df['Car Price'].loc[i])
```
