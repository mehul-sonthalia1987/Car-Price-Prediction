```python
import os,random, sys, time
#from urllib.parse import urlparse
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
brand_model=[]
variant=[]
year=[]
fuel=[]
km=[]
owner=[]
location=[]
price=[]
```

```python
#1. Ahmedabad
val = ['Ahmedabad']
```

```python
browser = webdriver.Chrome('chromedriver.exe')
browser.maximize_window()
browser.get('https://www.cars24.com/buy-used-cars/')
time.sleep(10)
select_manually= browser.find_element_by_xpath("/html/body/div[1]/div[11]/div/div/div/div[2]/div[2]/button")
select_manually.click()
time.sleep(10)
input_text = browser.find_element_by_xpath("/html/body/div[1]/div[11]/div/div/div/div[1]/div/div[2]/div/input")
input_text.send_keys(val)
loc_select = browser.find_element_by_xpath("//ul[@class='_16Bvy']/li")
loc_select.click()
time.sleep(10)
view_cars = browser.find_element_by_xpath("//span[@class='_1Aa4N']")
view_cars.click()
```

```python
for z in range(100):    
    time.sleep(2)
    if z==0:
        time.sleep(2)
        browser.execute_script("window.scrollTo(0,2000)")
        time.sleep(4)
        browser.execute_script("window.scrollTo(0, window.scrollY + 1000)")
    time.sleep(10)
    browser.execute_script("window.scrollTo(0, window.scrollY + 750)")
    if z in range(95,100,1):
        time.sleep(10)
        browser.execute_script("window.scrollTo(0, window.scrollY + 1500)")        
    print("counter=",z)
```

```python
details=[]
car_details = browser.find_elements_by_xpath("//div[@class='col-4']")
for i in car_details:
    details.append(i.text.split('\n'))    
len(details)
```

```python
for i in range(len(details)):
    if details[i]==['']:
        continue
    brand_model.append(details[i][0])
    variant.append(details[i][1])
    km.append(details[i][2])
    owner.append(details[i][3])    
    fuel.append(details[i][4])
    price.append(details[i][6])
    location.append(val)
    year.append('NaN')
    #print('counter',i)
```

```python
print(len(brand_model), len(variant), len(year), len(fuel), len(km), len(owner), len(price))
browser.close()
```
